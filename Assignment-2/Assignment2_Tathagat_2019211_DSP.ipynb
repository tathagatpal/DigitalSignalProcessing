```python
import numpy as np
import scipy as sc
import math
import cmath
import matplotlib.pyplot as plt

#Function implementing the filter
def notch(frequency):
    z=complex(math.cos(frequency),math.sin(frequency))
    h=np.asarray(cmath.polar((3*2**0.5 - 6*z**-1 + 3*2**0.5*z**-2)/(4*2**0.5 - 6*z**-1 + 2*2**0.5*z**-2)))
    return(h)

omega = np.linspace(0, math.pi, 10000)
H=np.zeros(shape=(2,10000))
for i in range(10000):
    y=notch(omega[i])
    H[0][i]=y[0]
    H[1][i]=y[1]

plt.plot(omega,H[0])
plt.axis([0, math.pi,-0.25, 1.25])
plt.hlines(1/math.sqrt(2), 0, math.pi, 'r', 'dotted')
plt.title('1a. Magnitude Response of the given Filter')
plt.xlabel('ω')
plt.ylabel('|H(ω)|')
plt.xticks(np.arange(0,5*math.pi/4,math.pi/4), ['0', 'π/4', 'π/2', '3π/4', 'π'])
plt.yticks(np.arange(-0.25,1.5,0.25))
plt.grid(linestyle='dashed')

plt.show()
```

```python
import numpy as np
import scipy as sc
from scipy import signal
import math
import cmath
import matplotlib.pyplot as plt

num = [-0.375, 0.25, 1]
den = [1, 0.25, -0.375]
omega, h = signal.freqz(num,den)
omega, d = signal.group_delay((num,den))
phasewrap = np.zeros(512)
for i in range(512):
    phasewrap[i]= cmath.phase(h[i])

plt.subplot(2,1,1)
plt.plot(omega,np.unwrap(phasewrap))
plt.axis([0, math.pi,-13*math.pi/4, math.pi/4])
plt.title('3b. Unwrapped phase')
plt.xlabel('ω')
plt.ylabel('|H(ω)|')
plt.xticks(np.arange(0,5*math.pi/4,math.pi/4), ['0', 'π/4', 'π/2', '3π/4', 'π'])
plt.yticks(np.arange(-2*math.pi,math.pi/2,1*math.pi/2,dtype=float), ['-2π', '-3π/2','-π','-π/2','0'])
plt.grid(linestyle='dashed')

plt.subplot(2,1,2)
plt.plot(omega,d)
plt.xlim([0, math.pi])
plt.title('3b. Group Delay')
plt.xlabel('ω')
plt.ylabel('τ(ω)')
plt.xticks(np.arange(0,5*math.pi/4,math.pi/4), ['0', 'π/4', 'π/2', '3π/4', 'π'])
plt.grid(linestyle='dashed')

plt.subplots_adjust(left=0.1,
                    bottom=1,
                    right=0.9,
                    top=1.5,
                    wspace=0.4,
                    hspace=1)

plt.show()
```
